```python
from espnet2.bin.tts_inference import Text2Speech
from espnet2.utils.types import str_or_none
import argparse, os, time, torch
import scipy.io.wavfile
import shutil
```

```python
tag = "../logs/kss/G_0.pth"
config = "./kss.yaml"
```

```python
print('start')
text2speech = Text2Speech.from_pretrained(
    model_file = str_or_none(tag),
    train_config = str_or_none(config),
)
print('end')
```

```python
import soundfile as sf
```

```python
wav = text2speech("안녕하세요. 저는 이주현입니다. 반갑습니다. 드디어 되는건가요? 내가 훈련시킨 모델! 데이터 적게")["wav"]
print("good")
```

```python
sf.write("out.wav", wav.numpy(), text2speech.fs, "PCM_16")
```

```python
from pydub import AudioSegment
```

```python
def convert_wav_to_mp3(wav_file, mp3_file):
    # WAV 파일을 로드합니다.
    sound = AudioSegment.from_wav(wav_file)
    
    # MP3 파일로 변환합니다.
    sound.export(mp3_file, format="mp3")
```

```python
# WAV 파일 경로와 MP3 파일 경로를 지정합니다.
wav_file = "out.wav"
mp3_file = "out.mp3"
```

```python
# WAV 파일을 MP3 파일로 변환합니다.
convert_wav_to_mp3(wav_file, mp3_file)
```
